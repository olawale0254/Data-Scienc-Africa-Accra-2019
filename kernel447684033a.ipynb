```python
# resize image and turn pixel values between 0 and 1
IMG_SHAPE = 80

# cell labels are 1 if bactarial leaf blight, 2 if brown spot, 3 if leaf smut.
disease_labels = []

# cells 
disease_images = []

#first six infected and unfected cells for visualization only
LeafBlast = []
BrownSpot = []
Healthy = []
Hispa = []

#listdir("../input/cell_images/cell_images/")
base_path = '../input/rice-diseases-image-dataset/labelledrice/Labelled/'

#get Bacterial_leaf_blight cells
leafblast_path = base_path +'LeafBlast/'
for file in listdir(leafblast_path):
    if file.endswith('.jpg'):
        file_path = leafblast_path + file
        image = imageio.imread(file_path)
        image = cv2.resize(image,(IMG_SHAPE,IMG_SHAPE)).astype('float32')/255.0
        LeafBlast.append(image)
        disease_images.append(image)
        disease_labels.append(1)

            
#get Brown_spot cells
BrownSpot_path = base_path +'BrownSpot/'
for file in listdir(BrownSpot_path):
    if file.endswith('.jpg'):
        file_path = BrownSpot_path + file
        image = imageio.imread(file_path) 
        image = cv2.resize(image,(IMG_SHAPE,IMG_SHAPE)).astype('float32')/255.0   
        BrownSpot.append(image)
        disease_images.append(image)
        disease_labels.append(2)

#get Leaf_smut cells
Healthy_path = base_path +'Healthy/'
for file in listdir(Healthy_path):
    if file.endswith('.jpg'):
        file_path = Healthy_path + file
        image = imageio.imread(file_path) 
        image = cv2.resize(image,(IMG_SHAPE,IMG_SHAPE)).astype('float32')/255.0   
        Healthy.append(image)
        disease_images.append(image)
        disease_labels.append(3)


#get Leaf_smut cells
Hispa_path = base_path +'Hispa/'
for file in listdir(Hispa_path):
    if file.endswith('.jpg'):
        file_path = Hispa_path + file
        image = imageio.imread(file_path) 
        image = cv2.resize(image,(IMG_SHAPE,IMG_SHAPE)).astype('float32')/255.0   
        Hispa.append(image)
        disease_images.append(image)
        disease_labels.append(4)
        
        
```
